```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import pandas as pd
import time
import seaborn as sns
from bs4 import BeautifulSoup
import requests
import csv
```

```python
url = "https://accesskent.com/Departments/Elections/Results/2018/nov2018-precinct.htm"
my_page = requests.get(url)
```

```python
# Use content 
kent_soup = BeautifulSoup(my_page.content, "html.parser")
```

```python
# Get div of interest
div = kent_soup.find('div', attrs={'class' : 'large-12 columns'})
```

```python
atags = div.find_all('a'); atags

# first three atags are useless for us 
atags = atags[3:]
```

```python
# Parse each url and get HTML
election_lst = []
for atag in atags: 
  # Initialize election_obj dictionary
    accesskent = 'https://accesskent.com/'

    election_obj = {} 
    election_obj['name'] = atag.text.strip()
    url = accesskent + atag.attrs['href']

    election_obj['url'] = url
    if(election_obj['name'] == "County Commissioner 3"): 
        election_obj['url'] = "https://www.accesskent.com/Departments/Elections/Results/2018/Nov/County%20Comm%203.htm"
    if(election_obj['name'] == "Kelloggsville Public Schools Bond Proposal"): 
        election_obj['url'] = "https://www.accesskent.com/Departments/Elections/Results/2018/Nov/Kelloggsville%20Public%20Schools%20Bond%20Prop.htm"
    election_lst.append(election_obj)  

    
```

```python
for election in election_lst: 
    name = election['name'].replace('/','_').replace(' ', '').replace("II.",'').replace("I.",'') + '.csv'
    election_page = requests.get(election['url'])
    page = BeautifulSoup(election_page.content, "html.parser")
    table = page.find("table")

    output_rows = []
    for table_row in table.findAll('tr'):
        columns = table_row.findAll('td')
        output_row = []
        for column in columns:
            output_row.append(column.text)
        output_rows.append(output_row)

    
    lst = list(range(1,len(output_rows),25))
    lst.remove(1)
    print(name)
    print(lst)
    print(len(output_rows))
    if(len(output_rows) == 127): 
        lst.remove(126)
        
    if(len(output_rows) == 260): 
        lst.remove(251)
        
    if(lst == [26]): 
        if(name not in ["76thStateRepresentative.csv", "77thStateRepresentative.csv", 
                        "86thStateRepresentative.csv"]):
            lst.remove(26)
            print(lst)
    
    
    for item in lst: 
        del output_rows[item]
    
    with open("data/" + name, 'w') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(output_rows)
    
```
